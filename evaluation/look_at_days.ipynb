```python
def get_to_plot_actions(actions):
    
    skeys = sorted(actions.keys())
    to_plot = []
    for k in skeys:
        to_plot.append(sum(actions[k])/len(actions[k]))
    return to_plot
def get_to_plot_regrets(actions):
    
    skeys = sorted(actions.keys())
    to_plot = []
    for k in skeys:
        to_plot.append(np.array(actions[k]).mean())
    return to_plot
```

```python
def adjust_to_day(steps):
    return steps.tz_localize(None).resample('30T').sum().fillna(0)
```

```python
def get_all_30_minutes_by_group(gid):
    gids = [k for k in groups if groups[k]==gid and k!=4]
    all_steps = {}
    for g in gids:
        steps = adjust_to_day(merged[g]['steps'])
        for i,s in steps.iteritems():
            time_stamp = '{}-{}'.format(i.hour,i.minute)
            if time_stamp not in all_steps:
                all_steps[time_stamp]=[]
            all_steps[time_stamp].append(s)
    return all_steps
```

```python
def get_mean_std(raw_array):
    arr = [s for s in raw_array]
    
    phi = np.sqrt(np.array(arr).std()**2+np.array(arr).mean()**2)
    
    normal_mean = math.log(np.array(arr).mean()**2/phi)
    normal_std = np.sqrt(math.log(phi**2/np.array(arr).mean()**2))
    return normal_mean
```
